**Load data**

```python
# %pip install opendatasets
import opendatasets as od 
import pandas 

od.download( 
	"https://www.kaggle.com/datasets/mohamedsaqibshouqi/2023-2024-nba-player-stats-playoffs") 
```

```python
import pandas as pd
file =('2023-2024-nba-player-stats-playoffs/NBA Stats 202324 All Stats.csv') 
data = pd.read_csv(file)
data
```

**Analyze data distribution**

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 6))
sns.countplot(data=data, y='POS', order=data['POS'].value_counts().index)
plt.title('Distribution of Different Positions')
plt.xlabel('Count')
plt.ylabel('Positions')
plt.show()
```

```python
data['AGE'] = round(data['AGE'])
plt.figure(figsize=(10, 6))
sns.histplot(data['AGE'], bins = 20, kde=True)
plt.title('Distribution of Age')
plt.xlabel('Count')
plt.ylabel('Age')
plt.show()
```

```python
fig, axs = plt.subplots(2, 3, figsize=(25, 15))
axs = axs.flatten()
sns.histplot(data['PPG'], bins=20,  kde=True, ax=axs[0])
axs[0].set_title('Distribution of Point per Game')

sns.histplot(data['APG'], bins=20,  kde=True, ax=axs[1])
axs[1].set_title('Distribution of Assist per Game')

sns.histplot(data['RPG'], bins=20,  kde=True, ax=axs[2])
axs[2].set_title('Distribution of Rebound per Game')

sns.histplot(data['BPG'], bins=20,  kde=True, ax=axs[3])
axs[3].set_title('Distribution of Blocks per Game')

sns.histplot(data['SPG'], bins=20,  kde=True, ax=axs[4])
axs[4].set_title('Distribution of Steals per Game')

sns.histplot(data['TPG'], bins=20,  kde=True, ax=axs[5])
axs[5].set_title('Distribution of Turnovers per Game')

plt.tight_layout()
plt.show()
```

**Analyze top 10 players**

The NBA 2024 MVP is Nikola Jokic. We will look at his performance in comparison to the top 10 players in each category

```python
mvp = data[data['NAME'] == 'Nikola Jokic']
mvp[['RANK','NAME','TEAM','POS','AGE','PPG','RPG','APG','SPG','BPG']]
```

```python
fig, axs = plt.subplots(1, 2, figsize=(18, 6))
top10_ppg = data[['RANK', 'NAME', 'PPG']].sort_values(by='PPG', ascending=False).head(10)
sns.barplot(x='PPG', y='NAME', data=top10_ppg, hue='NAME', palette='ocean', ax=axs[0])
axs[0].set_title("Top 10 Players by PPG")

sns.scatterplot(x='PPG', y='AGE', data=data, hue='POS', ax=axs[1])
axs[1].set_title("PPG vs Age")
axs[1].text(28.7,29,"Nikola Jokic") 
plt.show()
```

Jokic is ranked relatively high for PPG.

```python
fig, axs = plt.subplots(1, 2, figsize=(18, 6))
top10_rpg = data[['RANK', 'NAME', 'RPG']].sort_values(by='RPG', ascending=False).head(10)
sns.barplot(x='RPG', y='NAME', data=top10_rpg, hue='NAME', palette='ocean', ax=axs[0])
axs[0].set_title("Top 10 Players by RPG")

sns.scatterplot(x='RPG', y='AGE', data=data, hue='POS', ax=axs[1])
axs[1].set_title("RPG vs Age")
axs[1].text(13.4,29,"Nikola Jokic") 
plt.show()
```

Jokic is among the top 3 for RPG as a center. Josh Hart is the only guard in the top 10 with such high RPG.

```python
fig, axs = plt.subplots(1, 2, figsize=(18, 6))
top10_apg = data[['RANK', 'NAME', 'APG']].sort_values(by='APG', ascending=False).head(10)
sns.barplot(x='APG', y='NAME', data=top10_apg, hue='NAME', palette='ocean', ax=axs[0])
axs[0].set_title("Top 10 Players by APG")

sns.scatterplot(x='APG', y='AGE', data=data, hue='POS', ax=axs[1])
axs[1].set_title("APG vs Age")
axs[1].text(8.7,29,"Nikola Jokic")
axs[1].text(8.9,40,"LeBron James")
plt.show()
```

Despite the fact that assist are often done by guards, Jokic is ranked 2nd after LeBron James.

```python
fig, axs = plt.subplots(1, 2, figsize=(18, 6))
top10_spg = data[['RANK', 'NAME', 'SPG']].sort_values(by='SPG', ascending=False).head(10)
sns.barplot(x='SPG', y='NAME', data=top10_spg, hue='NAME', palette='ocean', ax=axs[0])
axs[0].set_title("Top 10 Players by SPG")

sns.scatterplot(x='SPG', y='AGE', data=data, hue='POS', ax=axs[1])
axs[1].set_title("SPG vs Age")
axs[1].text(1.4,29,"Nikola Jokic") 
axs[1].text(2.2,39,"LeBron James")
plt.show()
```

James is also leading when it comes to steals

```python
fig, axs = plt.subplots(1, 2, figsize=(18, 6))
top10_bpg = data[['RANK', 'NAME', 'BPG']].sort_values(by='BPG', ascending=False).head(10)
sns.barplot(x='BPG', y='NAME', data=top10_bpg, hue='NAME', palette='ocean', ax=axs[0])
axs[0].set_title("Top 10 Players by BPG")

sns.scatterplot(x='BPG', y='AGE', data=data, hue='POS', ax=axs[1])
axs[1].set_title("BPG vs Age")
axs[1].text(0.7,29,"Nikola Jokic")
plt.show()
```

Even with an impressive height of 6'11 (2.11m), blocks are not Jokic's strong suit
